```python
import os, glob, pandas as pd

input_folder = 'input_.csv_file'
output_folder = 'processed_.csv_file'
os.makedirs(output_folder, exist_ok=True)

all_files = glob.glob(os.path.join(input_folder, '*.csv'))
print(f"Found {len(all_files)} files to process in '{input_folder}'")

for file_path in all_files:
    base_name, ext = os.path.splitext(os.path.basename(file_path))
    output_file_name = f"{base_name}P{ext}"
    output_file_path = os.path.join(output_folder, output_file_name)

    if os.path.exists(output_file_path):
        # already processed
        continue

    skip = 0
    with open(file_path, 'r', encoding='utf-8') as f:
        for i, line in enumerate(f):
            if all(col in line for col in ['Date', 'Hour', 'Ontario Demand']):
                skip = i
                break

    df = pd.read_csv(file_path, skiprows=skip)
    required_columns = ['Date', 'Hour', 'Ontario Demand']
    df = df[[c for c in required_columns if c in df.columns]]

    df.to_csv(output_file_path, index=False)
    print(f"Processed and saved: {output_file_path}")

print("Data processing complete.")
print("Processed files are saved in:", os.path.abspath(output_folder))
```

```python
#csv file cleaner.ipynb

import os
import glob
import pandas as pd

input_folder = 'input_.csv_file'
output_folder = 'processed_.csv_file'

os.makedirs(output_folder, exist_ok=True)

all_files = glob.glob(os.path.join(input_folder, '*.csv'))
print(f"Found {len(all_files)} files to process in '{input_folder}'")

for file_path in all_files:
    skip = 0

    # Open the file to locate the real header row
    with open(file_path, 'r', encoding='utf-8') as f:
        for i, line in enumerate(f):
            if all(col in line for col in ['Date', 'Hour', 'Ontario Demand']):
                skip = i
                break

    df = pd.read_csv(file_path, skiprows=skip)

    required_columns = ['Date', 'Hour', 'Ontario Demand']
    df = df[[col for col in required_columns if col in df.columns]]

    base_name, ext = os.path.splitext(os.path.basename(file_path))
    file_name = f"{base_name}P{ext}"  # <-- now base_name exists

    output_file_path = os.path.join(output_folder, file_name)

    df.to_csv(output_file_path, index=False)

    print(f"Processed and saved: {output_file_path}")

print("Data processing complete.\n")

print("Processed files are saved in:")
print(os.path.abspath(output_folder))
```

```python
import os, glob, pandas as pd

input_folder = 'input_.csv_file'
output_folder = 'processed_.csv_file'
os.makedirs(output_folder, exist_ok=True)

all_files = glob.glob(os.path.join(input_folder, '*.csv'))
print(f"Found total {len(all_files)} files in '{input_folder}'")
print(f"{input_folder}: {os.path.abspath(input_folder)}")

already_processed = 0
for file_path in all_files:
    base_name, ext = os.path.splitext(os.path.basename(file_path))
    output_file_name = f"{base_name}.P{ext}"
    output_file_path = os.path.join(output_folder, output_file_name)

    if os.path.exists(output_file_path):
        already_processed += 1

print(f"Among these {len(all_files)} files, {already_processed} files are already processed.")

new_files = len(all_files) - already_processed
print(f"\nAnd,")
print(f"Found {new_files} new files to process in '{input_folder}'")

for file_path in all_files:
    base_name, ext = os.path.splitext(os.path.basename(file_path))
    output_file_name = f"{base_name}.P{ext}"
    output_file_path = os.path.join(output_folder, output_file_name)

    if os.path.exists(output_file_path):
        continue

    skip = 0
    with open(file_path, 'r', encoding='utf-8') as f:
        for i, line in enumerate(f):
            if all(col in line for col in ['Date', 'Hour', 'Ontario Demand']):
                skip = i
                break

    df = pd.read_csv(file_path, skiprows=skip)
    required_columns = ['Date', 'Hour', 'Ontario Demand']
    df = df[[c for c in required_columns if c in df.columns]]

    df.to_csv(output_file_path, index=False)
    print(f"Processed and saved: {os.path.join(output_folder, output_file_name)}")

print("\nData processing complete.")

print(f"All processed files are saved in: '{output_folder}'", os.path.abspath(output_folder))
```
